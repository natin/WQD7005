```python
#import related package
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
volume=pd.read_csv('Volume.csv')
```

```python
volume.head()
```

```python
volume.describe()
```

```python
volume.info()
```

```python
#make the dates as index
volume= volume.set_index(pd.to_datetime(volume['Dates']),inplace=False,drop=True)
volume = volume.drop('Dates',axis=1)
volume.head()
```

```python
#since we have 778 counter in klse
#we decide to filter it out by looking at the mean volume, 
#if the mean volume of each counter is below total volume, we drop
#as result we get 133 counter
#volume.describe()
mean=volume.mean()
mean.head()
```

```python
mean.mean()
```

```python
summation=volume.sum()
```

```python
summation=pd.DataFrame(summation)
summation.head()
```

```python
#to check how many counter that the volume is above the mean
meandf=pd.DataFrame(mean)
meandf[meandf[0]>1.414910e+06].head()
len(meandf[meandf[0]>1.414910e+06])
```

```python
active_counter=meandf[meandf[0]>1.414910e+06]
```

```python
fig = plt.figure(figsize=(30,22))
ax = plt.axes()
ax.plot(active_counter)
```

```python
active_counter.sort_values(by=0,ascending=False).head(10)
```
